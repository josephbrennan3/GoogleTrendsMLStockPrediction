```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score
from datetime import date
```

```python
df = pd.read_csv('/Users/josephbrennan/Desktop/Data/Merged/pson_merged.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df['Date'] = pd.to_datetime(df.Date)
df.index = pd.DatetimeIndex(df.Date)
df.drop('Date', axis=1, inplace=True)
df.head()
```

```python
#Convert to OneHotEncoder
df = pd.get_dummies(df)
```

```python
#Check for Null
df.isna().sum()
```

```python
y = df.iloc[:, 3]
X = df.iloc[:, [0, 1, 2, 4, 5, 6, 7, 8, 9, 10]]
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
#Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
```

```python
# Tuning of parameters for regression by cross-validation
K = 10 # Number of cross valiations
```

```python
parameters = [{'kernel': ['poly'], 'gamma': [1e-4, 1e-3, 0.01, 0.1, 0.2, 0.5, 0.6, 0.9],'C': [1, 10, 100, 1000, 10000]}]
svr = GridSearchCV(SVR(epsilon = 0.01), parameters, cv = K)
```

```python
svr.fit(X, y)
```

```python
svr.cv_results_
```

```python
results = pd.DataFrame(svr.cv_results_)
results[['param_C', 'param_gamma', 'param_kernel', 'rank_test_score']]
```

```python
print(results)
```
